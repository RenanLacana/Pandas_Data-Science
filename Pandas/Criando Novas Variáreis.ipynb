## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel_residencial.csv", sep=";")
```

```python
dados.head()
```

## Criando a Variável "Valor Bruto"

```python
dados["Valor Bruto"] = dados.Valor + dados.Condominio + dados.IPTU
dados
```

## Criando a Variável "Valor m2"

```python
dados["Valor m2"] = dados.Valor / dados.Area
dados
```

```python
dados["Valor m2"] = dados["Valor m2"].round(2)
```

```python
dados
```

## Criando a Variável "Valor Bruto m2"

```python
dados["Valor Bruto m2"] = (dados["Valor Bruto"] / dados.Area).round(2)
```

```python
dados
```

## Agregando tipos de Casa em uma Variável "Tipo Agregado"

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados.Tipo.apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

.apply() #aplica uma função em cada linha

## Excluindo Variáveis

Variável valor bruto e valor bruto m2 não são confiáveis porque existem muitos imóveis com valor de IPTU nulo. Foi decidido remover essas variáveis.

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux
```

```python
del dados_aux['Valor Bruto']
dados_aux
```

```python
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

```python
dados.drop(["Valor Bruto", "Valor Bruto m2"], axis = 1, inplace = True)
```

```python
dados
```

```python
dados.to_csv("aluguel_residencial.csv", sep=";", index=False)
```
